## Ecercício 1

Faça um programa que imprima a metade de todos os elementos de uma lista que contém alguns números quaisquer.

```python
# Código Exercício 1
lista = [1, 5, 8, 45, 62, 78, 92, 91, 12, 1, 0, 5]

for numero in lista:
  metade = numero/2
  print(metade)
```

## Exercício 2
Faça um programa que calcula e imprime na tela a soma dos itens de uma lista.

```python
# Código Exercício 2
lista = [1, 5, 8, 45, 62, 78, 92, 91, 12, 1, 0, 5]
soma = 0

for item in lista:
  soma = soma + item
  print(soma)
```

## Exercício 4
Faça um programa que imprima todos os números pares entre 1 e 50.


```python
# Código Exercício 4
numero = 1

while numero <= 50:
  if numero % 2 == 0:
    print(numero)
  numero += 1
```

## Exercício 5
Faça um programa que conte de 10 até 0 (de trás pra frente mesmo).

```python
# Código Exercício 5
numero = 10

while numero >= 0:
  print(numero)
  numero = numero - 1
```
