> Dica: Você pode usar um loop!

````python
produtos.append({"nome": "Mouse Gamer",
    "tipo": 2,
    "preço": 56.00})

produtos.append({"nome": "Mouse Pad",
    "tipo": 2,
    "preço": 32.90})

contador_codigo = 100

for x in produtos:
    x["codigo"] = contador_codigo
    contador_codigo += 1
````

5. Descobrimos que o melhor jeito de guardar as informações não é em listas, mas em um **dicionário de dicionários**, onde a **chave é o código do produto**. Usando loops faça essa **transformação até ficar no seguinte formato:**

```py
produtos = {
    100: {
        "nome": "Suporte de monitor",

    }
}
```



````python
novos_produtos = {}

for x in produtos:
    cod = x['codigo']
    x.pop('codigo')
    novos_produtos[cod] = x
````
