**Cenários requisitados:**

1. Se aumentarmos os valores de todos os produtos em 5%
2. Qual a média de preço após essa operação
3. Se diminuirmos os valores de todos os produtos em 10%
4. Qual a mediana de preço após essa operação

````python
# BASE DE VENDAS
base = pd.read_json("https://storage.googleapis.com/pedagogico/formacao-dados/tabela_vendas.json")

#Criando uma coluna com o total
base['total'] = base['quantidade'].mul(base['valor_produto'])

#Simulando cenários
##Aumento de 5%
base['total_mais_5p'] = base['total'].mul(1.05)
##Redução de 10%
base['total_menos_10p'] = base['total'].mul(0.9)

print(base[['total_mais_5p']].mean())
print(base[['total_menos_10p']].median())
````
