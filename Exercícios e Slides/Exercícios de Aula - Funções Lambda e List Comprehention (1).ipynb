##Exercício 1

Nos exercícios de Listas você fez a seguinte questão:

> Faça um programa que conta quantas letras *“a”* existem numa determinada palavra. Imprima o resultado na tela.

**Vamos repetir esse mesmo exercício elevando a complexidade juntando conceitos** que você já aprendeu até agora. Para resolver você deve utilizar os **seguintes requisitos:**
- O código precisa ser **envolvido em uma função `conta_letras`** para que possa ser reutilizado
- A função `conta_letras` precisa ter **2 argumentos**:
    1. O caractere que será contado
    2. A String que precisamos verificar
- Usar a técnica de **List Comprehension** para percorrer a String

````python
def conta_letras(letra, texto):
  lista = [1 if item == letra else 0 for item in texto]
  return(sum(lista))

conta_letras("x", "alo")
````

**Para acessar a base utilize o seguinte código:**


````python
# BASE DE CLIENTES
base = pd.read_csv("https://storage.googleapis.com/pedagogico/formacao-dados/tabela_clientes.csv")

base['cep_1dig'] = base['cep'].apply(lambda x: str(x)[0])

def de_para_regiao(regiao_cep, estado):
  regiao_cep = int(regiao_cep)

  if regiao_cep in [0,1,2,3]:
    return("Sudeste")
  elif regiao_cep in [8,9]:
    return("Sul")
  elif regiao_cep in [4,5,6] and estado not in ['Pará', 'Amazonas', 'Acre', 'Amapá' , 'Roraima']:
    return("Nordeste")
  elif estado in ['Pará', 'Amazonas', 'Acre', 'Amapá' , 'Roraima', 'Tocantins', 'Rondônia']:
    return('Norte')
  elif regiao_cep == 7 and estado not in ['Tocantins', 'Rondônia']:
    return("Centro-Oeste")

base['regiao'] = base.apply(lambda x: de_para_regiao(x['cep_1dig'], x['estado']), axis = 1)

base['regiao'].value_counts()
````
