```python
print('Functions Lambda')
import pandas as pd
carros = {
    'modelo': ['Maruti 800 AC','Maruti Wagon R LXI Minor','Hyundai Verna 1.6 SX', 'Datsun RediGO T Option', 'Honda Amaze VX i-DTEC', 'Maruti Alto LX BSIII', 'Hyundai Xcent 1.2 Kappa S', 'Tata Indigo Grand Petrol', 'Hyundai Creta 1.6 VTVT S', 'Maruti Celerio Green VXI', 'Chevrolet Sail 1.2 Base', 'Tata Indigo Grand Petrol', 'Toyota Corolla Altis 1.8 VL CVT'],
    'ano': [2007, 2007, 2012, 2017, 2014, 2007, 2016, 2014, 2015, 2017, 2015, 2014, 2018],
    'preço': [60000, 135000, 600000, 250000, 450000, 140000, 550000, 240000, 850000, 365000, 260000, 250000, 1650000],
    'combustivel': ['Petrol', 'Petrol', 'Diesel', 'Petrol', 'Diesel', 'Petrol', 'Petrol', 'Petrol', 'Petrol', 'CNG', 'Petrol', 'Petrol', 'Petrol']
}
tabela_carros = pd.DataFrame(carros)
tabela_carros
tabela_carros['descontos']=tabela_carros['preço'].apply(lambda x: x-(x*0.2))
tabela_carros['valor_desconto'] = tabela_carros['preço'].sub(tabela_carros['descontos'])
tabela_carros=tabela_carros.rename(columns={'preço':'preco_original','descontos':'valor_com_descontos','valor_desconto':'valor_do_desconto'})
tabela_carros['percentual']=tabela_carros['valor_com_descontos'].div(tabela_carros['preco_original']/100)
tabela_carros=tabela_carros.rename(columns={'percentual':'percentual_da_venda_%'})
tabela_carros.drop('percentual_da_venda_%',axis=1,inplace=True)
len(tabela_carros)
tabela_carros
```
