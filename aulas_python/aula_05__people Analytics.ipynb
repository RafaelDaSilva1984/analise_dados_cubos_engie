# Conceito - People Analytics
> **People Analytics** ou **análise de pessoas** é a abordagem orientada a dados que coleta, organiza e diagnostica dados sobre as equipes da empresa por meio de softwares e análises que cruzam as informações de diversas fontes — tudo para otimizar o gerenciamento de Recursos Humanos.

## Turnover

> Também conhecido como taxa de "Turnover" (do inglês) é o índice de rotatividade que simboliza a taxa média de saída de funcionários (demissões voluntárias e involuntárias) em relação ao número médio de funcionários de uma empresa em determinado espaço de tempo.

> O índice de rotatividade ideal varia de empresa para empresa mas, em geral, deve ser menos de 10% ao ano (1% ao mês).


## Turnover geral

 $$ Turnover geral = \dfrac{\dfrac{Numero \, de \, admissoes + Numero  \, de \, desligamentos}{2}}{Numero \, total \, de \, funcionarios} * 100$$


> Reflete a taxa de “movimentação” do número de funcionários dentro da empresa.

> Recomendada se você estiver querendo avaliar a rotatividade geral da empresa sem muita preocupação em extrair alguma informação qualitativa.


## Turnover de desligados
Para as pessoas desligadas da empresa, tem-se duas definições:
> **Desligados Ativos**: Quem pediu demissão;
> $$Turnover \, de \, desligados \, ativos = \dfrac{numero \, de \, desligados ativos(pediram \, demissao)}{todal \, de \, funcionarios} * 100$$ 
> **Desligados Passivos**: Quem foi demitido;
> $$Turnover \, de \, desligados \, passivos = \dfrac{numero \, de \, desligados \, passivos(demitidos)}{todal \, de \, funcionarios} * 100$$

> Se o turnover de **desligados passivos** é alto, significa que há um problema na contratação de funcionários. O processo de recrutamento e seleção muito provavelmente está deixando entrar na empresa pessoas sem fit cultural com a empresa ou que não atingem os níveis mínimos de performance nas avaliações de desempenho;

> Se turnover de **desligados ativos** é alto, a falha está na retenção de talentos. Faça uma avaliação de clima organizacional, procure por problemas no relacionamento com superiores, veja as condições de trabalho, ou seja, há algum problema sério na organização;


## Turnover de admissão
$$Turnover \, de \, admissão = \dfrac{numero \, de \, desligados \, com \, menos \, de \, 6 \, meses}{total \, de \, funcionarios} * 100 $$

> % de turnover para funcionários com até 6 meses;

> Um desligamento de uma pessoa tão recente indica uma falha grave no processos de seleção/recrutamento. Ou escolheu mal, ou não consegui transmitir para o candidato a real situação que ele encontraria, gerando um desalinhamento entre expectativa e realidade;

## Segmentar o turnover
É interessante analisar o turnover em diferentes níveis para se ter uma análise mais qualitativa, por exemplo:
> Por departamento;

> Por unidades de negócio;

> Por filiais;

> Por níveis hierárquicos;

> Por tempo de casa;

## Absenteísmo
$$= \dfrac{Numero \, de \, atrasos \, e \, faltas}{total \, de \, dias \, uteis} * 100 $$

> período em que os colaboradores se ausentam do trabalho, seja por atrasos, seja por faltas;

## Exercício 1
O time de RH de uma empresa quer ter uma noção do fluxo de funcionários durante o ano de 2016. Para isso, é necessário o cálculo do turnover geral. Qual o turnover geral de 2016?

```python
import pandas as pd
pd.set_option('display.max_columns',100) # mostra no máximo 100 colunas
```

```python
# Carregue os dados em um dataframe
people =pd.read_csv('https://storage.googleapis.com/pedagogico/formacao-dados/tabela_people_analytics.csv')
people
```

```python
# Verifique as 5 primeiras linhas do dataframe para entender um pouco mais sobre os dados
people.head()
```

```python
# Verifique as características estatíticas desse dataframe com o método .describe()
people.describe()
```

```python
# Crie uma seleçao que filtre no dataframe as pessoas desligadas, ou seja df['Termd'] == 1 e
# Coloque numa variavel chamada df_termination o dataframe filtrado
selecao_desligados = people['Termd'] == 1
people_termination = people[selecao_desligados] # people.loc[selecao_desligados]
people_termination.head()
```

```python
desligados_total = len(people_termination)
print(f'Desligados na totalidade = {desligados_total}')
```

```python
# encontre a quantidade de pessoas demitidas em 2016 utilizando filter e lambda function e list comprehension
#tratamento de dados 
# criado list comprehension linha a linha tabela(filtrado) da coluna desejada,usando  o split  separador '/', pegando último indice 2016
lista_desligamento = list(filter(lambda x: x == '2016', [linha.split('/')[-1] for linha in people_termination['DateofTermination']]))
#filtrar somente 2016 : filter + lambda function, criando uma lista somente o que for igual a 2016
#criar variavel que armazena tratamento desejado, para poder usar a mesma
#criar outra variável que armazena e manipula os dados necessários, usando o len() para saber quantidade de desligado em 2016
numero_desligados = len(lista_desligamento)
print(f'O número de desligados em 2016 é {numero_desligados}, de um total de {desligados_total} no geral da empresa.')
```

```python
# Encontre a quantidade de pessoas contratadas em 2016 utilizando filter e lambda function e list comprehension
lista_contratacao = list(filter( lambda x: x == '2016',[linha.split('/')[-1] for linha in people['DateofHire']]))
numero_contratados = len(lista_contratacao)
contratdos_total=len(people['DateofHire'])
print(f'O número de contratados em 2016 é {numero_contratados}, de um total de {contratdos_total} no geral da empresa.')

```

```python
# Calcule o turnover_geral = (((admissoes + desligamentos)/2) / total_funcionarios) * 100
total_funcionarios_2016 = 200
def calcula_turnover(numero_contratados, numero_desligados, total_funcionarios_2016):
    turnover = ((numero_contratados + numero_desligados)/2) / total_funcionarios_2016 * 100
    return turnover
result = calcula_turnover(numero_contratados, numero_desligados, total_funcionarios_2016)
print(f'Turnover igual: {result}')

```

# Exercicio 4
O pessoal de Employee Experience quer entender se existe algum impacto de algumas variáveis nas faltas e na satisfação dos colaboradores. São elas: estado civil, departamento, satisfação do colaborador. Eles fizeram algumas perguntas:


1. O departamento tem impacto na quantidade de faltas?
2. O departamento tem impacto na satisfação do coolaborador?
3. A participação em projetos tem impacto nas faltas?


### O estado civil solteiro tem impacto na quantidade de faltas? E o estado casado ?

```python
people['MaritalDesc'].unique()
#sem loop manual
solteiros = people[people['MaritalDesc']=='Single']
casados = people[people['MaritalDesc']=='Married']
print('Solteiros: ',solteiros['Absences'].mean())
print('Casados: ', casados['Absences'].mean(),'\n')

# Loop For
estados_civil = people['MaritalDesc'].unique() #variavel estados_civil, relaciona todos tipos da tabela[coluna]desejada
for status in estados_civil:
    df_est = people[people['MaritalDesc']==status]
    qtd_absences = df_est['Absences'].mean()
    
    print(status, ':', qtd_absences)
```

### O departamento do colaborador tem impacto na quantidade de faltas?

```python
people['Department'].unique()
#eliminar espaços vazios
people['Department_tratada']=people['Department'].apply(lambda x: x.strip())
people['Department_tratada'].unique()

#teste manual department x department
df_dep_ti = people[people['Department_tratada']=='IT/IS']
print('IT/IS: ',df_dep_ti['Absences'].mean())
df_dep_soft_eng = people[people['Department_tratada']=='Software Engineering']
print('Software Engineering: ',df_dep_soft_eng['Absences'].mean(),'\n')

# Loop For
# Lista única de departamentos
departmentos = people['Department_tratada'].unique()
# Loop sobre os departamentos
for department in departmentos:
    # Filtrar o DataFrame pelo departamento atual
    df_dep = people[people['Department_tratada'] == department]
    
    # Calcular a média de ausências para o departamento atual
    qtd_absences = df_dep['Absences'].mean()
    
    # Imprimir o resultado
    print('Faltas: ', department + ':', qtd_absences)
```

### O Departamento tem impacto na satisfação do coolaborador?

```python
people['Department_tratada']=people['Department'].apply(lambda x: x.strip())
people['Department_tratada'].unique()

for department in departmentos:
    # Filtrar o DataFrame pelo departamento atual
    df_dep = people[people['Department_tratada'] == department]
    # Calcular a média de Satisfação para o departamento atual
    qtd_satisfacao = df_dep['EmpSatisfaction'].mean()
    
    # Imprimir o resultado
    print('Satisfação: ',department + ':', qtd_satisfacao)

    
```

### A participação em projetos tem impacto nas faltas?

```python
people['Department_tratada']=people['Department'].apply(lambda x: x.strip())
people['Department_tratada'].unique()
for department in departmentos:
    # Filtrar o DataFrame pelo departamento atual
    df_dep = people[people['Department_tratada'] == department]
    # Calcular a média de Pparticipação para projetos especiais para o departamento atual
    qtd_engajamento = df_dep['SpecialProjectsCount'].mean()
    
    # Imprimir o resultado
    print('Projetos: ',department + ':', qtd_engajamento)
```
