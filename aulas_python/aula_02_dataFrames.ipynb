Calcular desconto de 20% para produtos com quantidade maior que 5 unidades
1- criar nova coluna que manipula quantidade * valor_produto
2- criar variável que executa a condição >= 5 desconto 20%
3- selecionar a coluna da tabela a condição e aplicar a condição
4- criar variaável que armazena a operação desejada

```python
# 1- criar nova coluna que manipula quantidade * valor_produto
base_calc['Valor x Quantidade']= base_calc['quantidade'].mul(base_calc['valor_produto'])
condicao_calcula=base_calc['quantidade']>= 5
base_calc['valor_produto'].loc[condicao_calcula] = base_calc['valor_produto'].loc[condicao_calcula]* 0.8
base_calc['novo_valor_qtd_x_valor']=base_calc['quantidade'].mul(base_calc['valor_produto'])
base_calc
base_calc['valor_produto'].mean()
print('Media Novos valores ',
    base_calc['novo_valor_qtd_x_valor'].mean(),
    'Media Antigos valores ',
    base_calc['Valor x Quantidade'].mean()
)
desvio = base_calc['Valor x Quantidade'].div(base_calc['novo_valor_qtd_x_valor'])
desvio
```

```python
ping.columns
# criar função para selecionar primerio digito da coluna, para usar como indice na tabela
def primeiro_digito(input):
    # var, convertida para str, selecionando indice [0], primeiro caracter
    primeiro = str(input)[0]
    return primeiro

#criar planilha auxiliar para armazenar datos de indice para cada coluna
auxiliar = pd.DataFrame()
# loop para passar em cada coluna da tabvela ping
for coluna in ping.columns:
    #na tabela auxiliar será criado coluna que receberá da tabela ping o primeiro digito da coluna
    auxiliar[coluna] = ping[coluna].apply(primeiro_digito)
# criar nova coluna na tab: auxiliar, nome id, com o primeiro caracteres das colunas, aplicando o axis=1, inserindo nas linhas ca coluna id
auxiliar['id']=auxiliar.apply(''.join, axis=1)
# juntar a coluna id da planilha auxiliar na planilha original ping
ping['id']= auxiliar['id']
```
